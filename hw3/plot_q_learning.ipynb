```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

with open('DQN-Lunar.pkl', 'rb') as a:
    data = pickle.loads(a.read())
time_step_a = data['Timestep']
mean_reward_a = data['mean']
best_reward_a = data['best']

with open('DDQNFalseLander_1e4.pkl', 'rb') as b:
    data = pickle.loads(b.read())
time_step_b = data['Timestep']
mean_reward_b = data['mean']
best_reward_b = data['best']

with open('DDQNFalseLander_lr2e3.pkl', 'rb') as c:
    data = pickle.loads(c.read())
time_step_c = data['Timestep']
mean_reward_c = data['mean']
best_reward_c = data['best']

with open('DDQNFalseLander_lr3e3.pkl', 'rb') as d:
    data = pickle.loads(d.read())
time_step_d = data['Timestep']
mean_reward_d = data['mean']
best_reward_d = data['best']


plt.figure()
plt.plot(time_step_a, mean_reward_a, color='green', linestyle = '-')
plt.plot(time_step_a, best_reward_a, color='green', linestyle = '--')

plt.plot(time_step_b, mean_reward_b, color='red', linestyle = '-')
plt.plot(time_step_b, best_reward_b, color='red', linestyle = '--')

plt.plot(time_step_c, mean_reward_c, color='blue', linestyle = '-')
plt.plot(time_step_c, best_reward_c, color='blue', linestyle = '--')

plt.plot(time_step_d, mean_reward_d, color='magenta', linestyle = '-')
plt.plot(time_step_d, best_reward_d, color='magenta', linestyle = '--')

plt.title('Vanilla Q-Learning Vs. Double Q-Learning', fontsize=11)
plt.xlabel('Timesteps')
plt.ylabel('Mean Episode Reward')
plt.legend(['Mean_DQN', 'Best Mean_DQN', 'Mean_DDQN1', 'Best Mean_DDQN1', 'Mean_DQN2', 'Best Mean_DQN2', 'Mean_DDQN3', 'Best Mean_DDQN3'])
ax = plt.gca()
ax.xaxis.get_major_formatter().set_powerlimits((0,0))
plt.show()
```
